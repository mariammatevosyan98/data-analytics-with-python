## Merge

```python
import pandas as pd
```

```python
df_orders = pd.read_csv('../data/raw/orders.csv')
df_orders.head()
```

```python
df_orders_products_train = pd.read_csv('../data/raw/order_products_train.csv')
df_orders_products_train.head()
```

```python
df_products = pd.read_csv("../data/raw/products.csv")
```

```python
df_departments = pd.read_csv('../data/processed/departments_t.csv')
```

```python
df_aisles = pd.read_csv('../data/raw/aisles.csv')
```

```python
df_customers = pd.read_csv('../data/raw/customers.csv')
df_customers.head()
```

```python
df_states= pd.read_csv('../data/raw/states.csv')
df_states.head()
```

```python
df_products_departments = pd.merge(df_products, df_departments, on="department_id", how="left")
```

```python
df_products_departments.head()
```

```python
df_products_departments.drop(columns=['department_id'], inplace=True)
```

```python
df_products_enriched = pd.merge(df_products_departments, df_aisles, on="aisle_id", how="left")    
```

```python
df_products_enriched.head()
```

```python
df_products_enriched.drop(columns=['aisle_id'], inplace=True)
```

```python
df_train_transactions = pd.merge(df_orders, df_orders_products_train, on="order_id", how="inner")
```

```python
df_train_transactions.head()
```

```python
df_train_analytic = pd.merge(df_train_transactions, df_products_enriched, on="product_id", how="left", indicator=True)
```

```python
df_train_analytic.head()
```

```python
df_train_analytic.drop(columns=['eval_set','product_id','_merge'], inplace=True)
```

```python
df_train_analytic.head()
```

```python
df_train_analytic.shape
```

```python
df_customers = df_customers.rename(columns={'STATE': 'state'})
```

```python
df_customers_states = pd.merge(df_customers, df_states, on="state", how="left")
```

```python
df_customers_states.head()
```

```python
df_train_analytic_final = pd.merge(df_train_analytic, df_customers_states, on="user_id", how="left")
```

```python
df_train_analytic_final.head()
```

```python
df_train_analytic_final.drop(columns=['Unnamed: 0_y','user_id'], inplace=True)
```

```python
df_train_analytic_final.head()
```
