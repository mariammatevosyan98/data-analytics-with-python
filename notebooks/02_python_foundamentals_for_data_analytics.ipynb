### Part 3

```python
“Premium” if revenue > 250
“Standard” if 100 < revenue ≤ 250
“Low” otherwise
```

```python
sales = [20, 80, 100, 130, 200, 250, 300, 350]
segments = []

for revenue in sales:

    if revenue > 250:
        segments.append("Premium")
    elif 100 < revenue <= 250:
        segments.append("Standard")
    else:
        segments.append("Low")

segments
    
```

```python
premium = segments.count("Premium")
standard = segments.count("Standard")
low = segments.count("Low")

print(f"Premium: {premium} \nStandard: {standard} \nLow: {low}")
```
